dropping nulls

```python
dfc=dfn.dropna() #dropping all the rows with nulls
dfc
```

```python
dfc1=dfn.fillna(5)  #null values are replaced with specified value
dfc1
```

cleaning with mean

```python
m=dfn['SEM2'].mean()     #mean of sem2
print(m)
```

```python
dfc2=dfn.fillna(m)
print(dfc2)
```

```python
dropped=dfc2.drop_duplicates()    #duplicates will be remmoved
print(dropped)
```
