```python
import pandas as pd
```

```python
df=pd.read_csv('/content/demodt.txt')
df
```

```python
from sklearn.tree import DecisionTreeClassifier
LR=DecisionTreeClassifier()
target=df.Good
feat_list=['Literacy','Cleanliness','Crime_Rate']
feat=df[feat_list]
LR.fit(feat,target)
```

```python
pred=LR.predict([[90,90,78]])
print(pred)
```

```python
if(pred==1):
  print('Good')
else:
  print('Bad')
```

```python
Literacy=int(input("enter literacy number"))
Cleanliness=int(input("enter cleanliness number"))
Crime_Rate=int(input("enter crime rate number"))
pred=LR.predict([[Literacy,Cleanliness,Crime_Rate]])
print(pred)
if(pred==1):
  print('Good')
else:
  print('Bad')
```
